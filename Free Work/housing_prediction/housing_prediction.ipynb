<h4>Importing dependencies</h4>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

<h4>Reading dataset</h4>

```python
data = pd.read_csv('Data/Housing.csv')
print(data)
```

<h4>Assigning Features and Labels of the project</h4>

```python
features = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']
labels = 'price'
```

<h4>Creating X and Y</h4>

```python
x = data[features]
y = data[labels]

print(x)
print(y)
```

<h4>Train Test Split</h4>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
display(x_train, x_test)
```

<h4>Model Initialization</h4>

```python
model = LinearRegression()
```

<h4>Changing the string values into number values (integer or float)</h4>

```python
print(x_train.columns)
columns_to_encode = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
existing_columns = [col for col in columns_to_encode if col in x_train.columns]
x_train = pd.get_dummies(x_train, columns= existing_columns) #One hot Encoding for string values
x_test = x_test.reindex(columns=x_train.columns, fill_value=0) #some category or column are absent after encoding the trained data
```

```python

display(x_train)
```

<h4>Model Training</h4>

```python
model.fit(x_train, y_train)
```

<h4>Predicting House Price Using Features</h4>

```python
y_predict = model.predict(x_test)
display(y_predict.tolist(), y_test.tolist())
```

<h4>Mean Squared Error and R-Squared </h4>

```python
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print(f"Mean Squared Error : {mse}")
print(f"R-Squared Error : {r2}")
```

<h4>Display Slope and Intercept</h4>

```python
print(f"Slope (m): {model.coef_}")
print(f"Y-intercept (c): {model.intercept_}")
```

<h4>Scatter plot of the project </h4>

```python
plt.scatter(y_test, y_predict)
plt.title('Scatter plot of Y_test Vs Y_predict :')
plt.xlabel('Y_test')
plt.ylabel('Y_predict')
```
